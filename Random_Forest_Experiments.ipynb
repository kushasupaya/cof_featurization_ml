```python
import numpy as np
import pandas as pd
import sklearn.linear_model as sl
import sklearn.preprocessing as sp
import sklearn.decomposition as sd
import sklearn.model_selection as sm
import sklearn.ensemble as se
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, PowerTransformer, MinMaxScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error, r2_score

features = pd.read_csv('/content/features_5000 (1).csv')
properties = pd.read_csv('/content/properties_5000_cofs.csv', nrows=5000)
```

```python
X = features.drop(features.columns[[0]], axis =1 )
Y = properties[[' deliverable capacity [v STP/v]',' heat desorption low P [kJ/mol]',' absolute methane uptake low P [mol/kg]']]
X.head()
```

```python
x_train, x_test, y_train, y_test = sm.train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, PowerTransformer, MinMaxScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error, r2_score
#1 try

numeric_transformer1 =  Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler',StandardScaler()),
    ('poly',PolynomialFeatures(degree=1)),
])
numeric_features =X.columns
preprocessor1 = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer1, numeric_features)
])

rg = Pipeline(
    steps=[("preprocessor", preprocessor1), ("regressor", RandomForestRegressor())]
)

params = {

            "regressor__max_features"      : ["sqrt", "log2",None],
            "regressor__max_depth"         : [None,10,20,30,50],
            "regressor__min_samples_split" : [2,5,10],
            "regressor__min_samples_leaf"  : [1,2,4,8],
            "regressor__bootstrap": [True]
            }
```

```python
gp = sm.GridSearchCV(rg,param_grid=params)
gp.fit(x_train,y_train)
best_p = gp.best_params_
y_pred = gp.predict(x_test)
print("Tr:{}".format(np.mean(gp.predict(x_train)-y_train)**2))
print("Te:{}".format(np.mean(y_pred-y_test)**2))
print(mean_squared_error(y_test, y_pred))

print(gp.score(x_test,y_test))
print(gp.r2_score(y_test, y_pred))
```
